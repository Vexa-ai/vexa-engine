```python
from psql_helpers import get_session
from sqlalchemy import select
from psql_models import Thread
```

```python
async def debug_threads():
    async with get_session() as session:
        # Get all threads for user
        all_threads = await session.execute(
            select(Thread)
        )
        threads = all_threads.scalars().all()
        
        print(f"Total threads: {len(threads)}")
        print("\nThread details:")
        for t in threads:
            print(f"Thread ID: {t.thread_id}")
            print(f"Content ID: {t.content_id}")
            print(f"Thread name: {t.thread_name}")
            print("---")
```

```python
await debug_threads()
```

```python
from IPython.display import clear_output, display, Markdown
from typing import Optional, List
from functools import wraps
from helpers import jupyter_stream_output

@jupyter_stream_output(markdown=True)
async def stream_chat(
    client,
    query: str,
    meeting_id: Optional[str] = None,
    entities: Optional[List[str]] = None,
    thread_id: Optional[str] = None,
    model: str = "gpt-4o-mini",
    temperature: float = 0.7
):
    response = client.chat(
        query=query,
        meeting_id=meeting_id,
        entities=entities,
        thread_id=thread_id,
        model=model,
        temperature=temperature
    )
    yield response
```
