```python
from qdrant_search import QdrantSearchEngine
import pandas as pd
from psql_models import get_session
```

```python
# Initialize and setup
search_engine = QdrantSearchEngine()
```

```python

await search_engine.sync_from_postgres(get_session)

```

```python
pd.options.display.max_colwidth = 100
```

```python
# Search across meetings
results = await search_engine.search(
    query_text="alexey",
    limit=200,
    min_score=0.49
)
```

```python
speaker_results = await search_engine.search_by_speaker(
    speaker_query="rick",
    limit=200,
    min_score=0.49
)
```
